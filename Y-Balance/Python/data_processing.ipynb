```python
import os
import pandas as pd
import numpy as np
import os
import glob
import warnings
warnings.filterwarnings('ignore')

mo_cols = ['CentreOfMass',
'LAJC','LAnkleAngles',
'LEJC','LElbowAngles',
'LHJC','LHipAngles',
'LKJC','LKneeAngles',
'LPelvisAngles','LSpineAngles','LThoraxAngles',
'LSJC','LShoulderAngles',
'LWJC','LWristAngles',
'RAJC','RAnkleAngles',
'REJC','RElbowAngles',
'RHJC','RHipAngles',
'RKJC','RKneeAngles',
'RPelvisAngles','RSpineAngles','RThoraxAngles',
'RSJC','RShoulderAngles',
'RWJC','RWristAngles',]
```

```python
[basedir, Dir] = os.path.split(os.getcwd())
DIR = os.path.join(basedir, 'Data')
FILES = glob.glob(DIR +'/'+'*.csv')
store_dir = os.path.join(basedir, "Concat")

file = FILES[0]

for file in FILES:
    df = pd.read_csv(file, skiprows=2)
    mo_idx, trj_idx = np.where(df['Unnamed: 0'] == 'Frame')[0]

    mo = df.iloc[:trj_idx-3]
    trj = pd.read_csv(file, skiprows=trj_idx+3)


    mo_newcols = []
    mo_use_idx = []
    for idx, c in enumerate(mo.columns):
        name, col = c.split(':')
        if col in mo_cols:
            mo_use_idx.append(idx)
            mo_use_idx.append(idx+1)
            mo_use_idx.append(idx+2)
            col1 = col
            mo_newcols.append(f"{col1}_X")
            mo_newcols.append(f"{col1}_Y")
            mo_newcols.append(f"{col1}_Z")
        elif col in [' 0']:
            mo_use_idx.append(idx)
            mo_newcols.append('Frame')
        elif col in [' 1']:
            mo_use_idx.append(idx)
            mo_newcols.append('Sub Frame')
            
    mo = mo.iloc[2:,mo_use_idx].reset_index(drop=True)
    mo.columns = mo_newcols


    trj_newcols = []
    trj_use_idx = []
    for idx, c in enumerate(trj.columns):
        name, col = c.split(':')
        if "Unnamed" not in name:
            trj_use_idx.append(idx)
            trj_use_idx.append(idx+1)
            trj_use_idx.append(idx+2)
            col1 = col
            trj_newcols.append(f"{col1}_X")
            trj_newcols.append(f"{col1}_Y")
            trj_newcols.append(f"{col1}_Z")

    trj = trj.iloc[2:, trj_use_idx].reset_index(drop=True)
    trj.columns = trj_newcols

    concat_df = pd.concat([mo, trj], axis=1)
    del concat_df['Sub Frame']
    concat_df.to_csv(os.path.join(store_dir, os.path.basename(file)), index=False)
```
