```python
import sys
from Functions import Animated
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython import display
from Functions import Animated
import os
import pandas as pd
import glob

speed = 2
axis = 'z'
connections = [
    ('LAJC','LKJC'),('LKJC','LHJC'),('LHJC','RHJC'),('RHJC','RKJC'),('RAJC','RKJC'),
    ('LWJC','LEJC'),('LEJC','LSJC'),('LSJC','RSJC'),('RSJC','REJC'),('REJC','RWJC'),
]

[basedir, Dir] = os.path.split(os.getcwd())
DIR = os.path.join(basedir, 'Concat')
FILES = sorted(glob.glob(DIR +'/'+'*.csv'))[::3]

file = FILES[0]

usecols = [
'LAJC','LEJC','LHJC','LKJC','LSJC','LWJC',
'RAJC','REJC','RHJC','RKJC','RSJC','RWJC',
]
```

```python
for file in FILES:
    df = pd.read_csv(file)
    cols = []
    for i in df.columns:
        if i.split('_')[0] in usecols:
            cols.append(i)
    
    fname = os.path.basename(os.path.splitext(file)[0])
    markers = {}
    for i in range(0,len(cols),3):
        markers[cols[i].split('_')[0]] = df[cols[i:i+3]].values.T

    anim = Animated.Animate_3d(markers,speed,40,axis,fname,60,connections)
    video = anim.to_html5_video()
    html = display.HTML(video)
    display.display(html)
    anim.save("Videos/animation.mp4", fps=30)
    plt.close()
```
